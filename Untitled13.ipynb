```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import SimpleRNN, Dense
```

```python
df=pd.read_csv('birth_rate_rural.csv')
df.head(50)
```

```python
df.isnull().sum()
```

```python
df['Birth Rate'].fillna(method='bfill', inplace=True)
```

```python
birth_rates = df['Birth rate rural'].values.reshape(-1, 1)

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
birth_rates_scaled = scaler.fit_transform(birth_rates)

# Define parameters
time_steps = 5  # Number of time steps
forecast_period = 10  # Number of years to forecast

# Create dataset with time steps
X, Y = [], []
for i in range(len(birth_rates_scaled) - time_steps):
    X.append(birth_rates_scaled[i:(i + time_steps), 0])
    Y.append(birth_rates_scaled[i + time_steps, 0])
X, Y = np.array(X), np.array(Y)

# Reshape input data to be 3D [samples, time steps, features]
X = np.reshape(X, (X.shape[0], X.shape[1], 1))

# Build Simple RNN model
model = Sequential()
model.add(SimpleRNN(units=50, input_shape=(X.shape[1], 1)))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X, Y, epochs=100, batch_size=32)

# Make predictions
predictions = model.predict(X)

# Inverse transform predictions
predictions = scaler.inverse_transform(predictions)

# Plot actual vs predicted values
plt.figure(figsize=(10, 6))
plt.plot(df['Year'][time_steps:], birth_rates[time_steps:], label='Actual')
plt.plot(df['Year'][time_steps:], predictions, label='Predicted')
plt.xlabel('Year')
plt.ylabel('Birth Rate')
plt.title('Actual vs Predicted Birth Rates')
plt.legend()
plt.show()

# Forecast next 10 years
forecast = []
inputs = birth_rates_scaled[-time_steps:].reshape(1, time_steps, 1)

for i in range(forecast_period):
    prediction = model.predict(inputs)
    forecast.append(prediction[0, 0])
    inputs = np.append(inputs[:, 1:, :], prediction.reshape(1, 1, 1), axis=1)

# Inverse transform forecasted values
forecast = scaler.inverse_transform(np.array(forecast).reshape(-1, 1))

# Plot historical birth rates and forecasted birth rates
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], birth_rates, label='Historical')
plt.plot(np.arange(2021, 2021 + forecast_period), forecast, label='Forecasted')
plt.xlabel('Year')
plt.ylabel('Birth Rate')
plt.title('Historical vs Forecasted Birth Rates')
plt.legend()
plt.show()

# Create DataFrame for forecasted values
forecasted_years = np.arange(2021, 2021 + forecast_period)
forecasted_df = pd.DataFrame({
    'Year': forecasted_years,
    'Forecasted Birth Rate': forecast.flatten()
})

print(forecasted_df)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Calculate RMSE and MSE
actual_values = df['Birth rate rural'].values[-10:]
rmse = np.sqrt(mean_squared_error(actual_values, forecast.flatten()))
mse = mean_squared_error(actual_values, forecast.flatten())

# Print the results
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Squared Error (MSE):", mse)

```
