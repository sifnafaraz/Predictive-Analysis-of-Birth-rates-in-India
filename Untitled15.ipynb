```python
from statsmodels.tsa.stattools import kpss
import pandas as pd

# Function to perform KPSS test and difference data if non-stationary
def make_stationary(data):
    # Perform KPSS test
    kpss_statistic, p_value, lags, critical_values = kpss(data)

    # Print KPSS statistic and p-value
    print("KPSS Statistic:", kpss_statistic)
    print("p-value:", p_value)

    # Check if the data is non-stationary
    if p_value < 0.05:
        print("Data is non-stationary (reject null hypothesis of stationarity)")

        # Difference the data
        differenced_data = data.diff().dropna()
        print("Data differenced successfully.")

        # Recursively check for stationarity
        make_stationary(differenced_data)
    else:
        print("Data is stationary (fail to reject null hypothesis of stationarity)")
        return data

# Example usage
# Assuming 'data' is your non-stationary time series data stored in a Pandas Series
# Replace 'data' with your actual time series data
# Example:
# data = pd.Series([10, 15, 20, 25, 30, 35, 40, 45, 50])
# where the values represent a non-stationary time series

# Make the data stationary
stationary_data = make_stationary(data)
```

```python
import itertools
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 4) for x in list(itertools.product(p, d, q))]

print('Examples of parameter combinatins for seasonal ARIMA')
print('SARIMAX:{}x{}'.format(pdq[1],seasonal_pdq[1]))
print('SARIMAX:{}x{}'.format(pdq[1],seasonal_pdq[2]))
print('SARIMAX:{}x{}'.format(pdq[2],seasonal_pdq[3]))
print('SARIMAX:{}x{}'.format(pdq[2],seasonal_pdq[4]))
```

```python
import warnings
import pandas as pd
import statsmodels.api as sm

warnings.filterwarnings("ignore")

data1 = data['Birth rate urban']

for param in pdq:
    for param_seasonal in seasonal_pdq:
        model = sm.tsa.statespace.SARIMAX(data1,
                                           order=param,
                                           seasonal_order=param_seasonal,
                                           enforce_stationarity=False,
                                           enforce_invertibility=False)
        try:
            results = model.fit()
            print(f'ARIMA{param}x{param_seasonal}4 - AIC:{results.aic}')
        except Exception as e:
            print(f'Error for ARIMA{param}x{param_seasonal}4:{str(e)}')
            continue
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
sarima_model = SARIMAX(data['Birth rate urban'], order=(0, 1, 1), seasonal_order=(1, 1, 1, 4))
sarima_result = sarima_model.fit()
```

```python
print(sarima_result.summary())
```

```python
Best_model = sm.tsa.statespace.SARIMAX(data,
                                  order=(1, 0, 1),
                                  seasonal_order=(1, 0, 0, 4),
                                  enforce_stationarity=False,
                                  enforce_invertibility=False)
results =Best_model.fit()
print(results.summary().tables[1])
```

```python
import matplotlib.pyplot as plt
results.plot_diagnostics(figsize=(15, 12))
plt.show()
```

```python
pred = results.get_prediction(start=pd.to_datetime('01-01-1971'), dynamic=False)
pred_ci = pred.conf_int()
pred_ci
```

```python
y_forecasted = pred.predicted_mean
y_truth = data['01-01-1976':]
y_truth['Pred_Birth Rate_']=y_forecasted
y_truth
```

```python
import matplotlib.pyplot as plt

# Plotting the actual and predicted values
plt.figure(figsize=(10, 6))

# Plot actual values
plt.plot(y_truth.index, y_truth['Birth rate urban'], label='Actual', color='blue')

# Plot predicted values
plt.plot(y_truth.index, y_truth['Pred_Birth Rate_'], label='Predicted Birth Rate', color='red')

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Birth Rate')  # Corrected typo: 'ylabe' to 'ylabel'
plt.title('Actual vs Predicted Birth Rate')
plt.legend()

# Display the plot
plt.show()
```

```python
mse = ((y_truth['Birth rate urban'] - y_truth['Pred_Birth Rate_'])**2).mean()
```

```python
print('Mean Squared Error =',mse)
```

```python
rmse = np.sqrt(mse)
print('Root Mean Squared Error = ',rmse)
```
